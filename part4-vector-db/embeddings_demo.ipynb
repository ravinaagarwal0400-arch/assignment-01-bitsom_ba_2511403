```python
# Install & Import

!pip install sentence-transformers seaborn matplotlib

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sentences (10 total, 3 topics)

sentences = [
    # Cricket
    "The batsman scored a century in the final match",
    "The bowler delivered a perfect yorker to take the wicket",
    "The team won the match by chasing a huge target",

    # Cooking
    "The chef prepared a delicious pasta with fresh ingredients",
    "Baking a cake requires precise measurements and timing",
    "The recipe includes spices, herbs, and slow cooking techniques",

    # Cybersecurity
    "Strong passwords are essential to prevent unauthorized access",
    "Encryption protects sensitive data from cyber attacks",
    "Firewalls help secure networks from malicious traffic",
    "Phishing attacks trick users into revealing confidential information"
]
```

```python
# Generate Embeddings

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(sentences)

print("Embedding shape:", embeddings.shape)
```

```python
# Cosine Similarity Matrix

similarity_matrix = cosine_similarity(embeddings)

plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Cosine Similarity Heatmap")
plt.show()
```

```python
# Query Similarity Search

query = "The bowler took three wickets in one over"

query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 indices
top_indices = np.argsort(similarities)[-2:][::-1]

print("Query:", query)
print("\nTop 2 similar sentences:\n")

for idx in top_indices:
    print(f"Sentence: {sentences[idx]}")
    print(f"Similarity Score: {similarities[idx]:.4f}")
    print()
```
